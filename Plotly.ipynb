```python
z = [[.1, .3, .5, .7, .9],
     [1, .8, .6, .4, .2],
     [.2, 0, .5, .7, .9],
     [.9, .8, .4, .2, 0],
     [.3, .4, .5, .7, 100]]

colorscale = [[0, 'navy'],[1,'red']]
fon_colors = ["yellow", 'orange']
fig = ff.create_annotated_heatmap(z, colorscale=colorscale, font_colors=fon_colors)
fig.show()
```
